protopunica is a python library thet implements probalistic models

```python
smoking= Node(DiscreteDistribution({"high smoking":0.7,"low smoking":0.3}),name="smoking")
```

smoking is root node and is not having any parent nodes smoking is a random variables that takes two values high smoking and low smoking

```python
asbes_consum=Node(DiscreteDistribution({"high cons":0.3,"low cons":0.7}),name="asbes_consum")
```

asbestos consumption is also a root node which is not having 
any parent and has 2 values highconsumption and low consumption

```python
cancer=Node(ConditionalProbabilityTable([
    ["high smoking","high cons","POS",0.4],
    ["high smoking","high cons","Neg",0.6],
    ["high smoking","low cons","POS",0.3],
    ["high smoking","low cons","Neg",0.7],
    ["low smoking","low cons","POS",0.1],
    ["low smoking","low cons","Neg",0.9],
    ["low smoking","high cons","POS",0.02],
    ["low smoking","high cons","Neg",0.98]
],[smoking.distribution,asbes_consum.distribution]),
     name="cancer")
     
```

cancer is a random variable which conditionally depended on smoking and asbestos cancer takes two values positive and negative

```python
scan=Node(ConditionalProbabilityTable([
    ["POS","scan_POS",0.8],
    ["POS","scan_Neg",0.2],
    ["Neg","scan_POS",0.1],
    ["Neg","scan_Neg",0.9],
],[cancer.distribution]),name="scan")  
```

scan(x-ray)is conditionally dependent on cancer if there is a chance thay thre x-ray might revel it.it takes two values positive and negative

```python
blood_vomiting=Node(ConditionalProbabilityTable([
    ["POS","b.v_POS",0.7],
    ["POS","b.v_Neg",0.3],
    ["Neg","b.v_POS",0.2],
    ["Neg","b.v_Neg",0.8]
],[cancer.distribution]),name="blood_vomiting")    
```

patient suffers from blood vomitings blood vomiting takes two values B.vops(yes he has blood vomiting)
and B.V Neg(No he is not having any blood vomiting)

# Model bayesian network

```python
model=BayesianNetwork()
```

# Add all the nodes for the bayesian network

```python
model.add_states(smoking,asbes_consum,cancer,scan,blood_vomiting)
```

```python
model.add_edge(smoking,cancer)
model.add_edge(asbes_consum,cancer)
model.add_edge(cancer,scan)
model.add_edge(cancer,blood_vomiting)
```

```python
model.bake()
```

```python
model
```

```python
probability=model.probability([["low smoking","low cons","POS","scan_POS","b.v_POS"]])
```

```python
probability
```

```python
print(model.predict([["low smoking","low cons","POS","scan_POS","b.v_POS"]]))
```

```python
predictions=model.predict_proba({"blood vomiting":"b.v_POS"})
```
